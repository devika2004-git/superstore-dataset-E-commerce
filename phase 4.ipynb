step 3- DATA STANDARDIZATION

1. Standardizing Date Columns

Convert Order Date and Ship Date to datetime format.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

2. Standardizing Text Columns

Remove extra spaces and make text formatting consistent.

```python
text_columns = ['Customer Name', 'City', 'State',
                'Category', 'Sub-Category', 'Region']

for col in text_columns:
    df[col] = df[col].str.strip()
```

3. Standardizing Numerical Columns

Scale numerical features such as Sales, Quantity, Discount, and Profit.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Sales', 'Quantity', 'Discount', 'Profit']] = scaler.fit_transform(
    df[['Sales', 'Quantity', 'Discount', 'Profit']]
)
```

4. Standardizing Column Names

Convert column names into a consistent format.

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

step 4-CREATING DERIVED COLUMNS

Derived columns are new columns created from existing data to generate additional insights.

1. Shipping Days

Calculate the number of days taken to ship an order.

```python
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

2. Order Year

Extract the year from the order date.

```python
df['Order Year'] = df['Order Date'].dt.year
```

3. Profit Margin

Calculate the profit percentage for each transaction.

```python
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100
```

CONCLUSION:-three new columns are added.shipping days,order year,profit margin

PHASE4- VISUALIZATIONS

1. Bar Plot – Sales by Category
   

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("superstore.xlsx")

plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

The technology category with the highest sales contributes most to the company's revenue, while the office supplies category with the lowest sales may require improved marketing or product strategies

2.Line Chart – Sales Trend Over Time

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

Sales fluctuate over time, showing seasonal patterns and periods of growth or decline. Peak months indicate high customer demand.

3. Pie Chart – Sales Contribution by Region

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(7,7))
plt.pie(region_sales,
        labels=region_sales.index,
        autopct='%1.1f%%')
plt.title('Sales Contribution by Region')
plt.show()
```

Shows each region's contribution to total sales.west regions contribute a larger share of total sales than others, indicating stronger market presence and customer demand in those areas.

4. Histogram – Distribution of Sales

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

Reveals whether sales values are normally distributed or skewed.Most orders fall within a certain sales range, while a few high-value orders create a right-skewed distribution.

Box Plot – Profit by Category

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Category', y='Profit', data=df)
plt.title('Profit Distribution by Category')
plt.show()
```

Detects outliers and profit variations across categories.Profit varies significantly across categories, and the presence of outliers suggests some products generate exceptionally high or low profits.Here we can clearly says that technology gives higher profits

6. Scatter Plot – Sales vs Profit

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit',
                hue='Category',
                data=df)
plt.title('Sales vs Profit')
plt.show()
```

Examines the relationship between sales and profitability.Higher sales do not always result in higher profits. Some high-sales transactions may yield low profits due to discounts or costs.

7. Heatmap – Correlation Analysis

```python
numeric_cols = df[['Sales','Profit','Quantity','Discount']]

plt.figure(figsize=(8,5))
sns.heatmap(numeric_cols.corr(),
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Shows how numerical variables are related.Sales, Profit, Quantity, and Discount exhibit varying degrees of correlation. Discounts often show a negative relationship with profit.

8. Count Plot – Orders by Ship Mode

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Ship Mode', data=df)
plt.title('Orders by Ship Mode')
plt.xticks(rotation=15)
plt.show()
```

Identifies the most frequently used shipping method.The most frequently used shipping mode reflects customer preference and operational efficiency, while less-used modes may indicate higher costs or lower demand.

9. Top 10 Sub-Categories by Sales

```python
top_sub = df.groupby('Sub-Category')['Sales'].sum()\
            .sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_sub.plot(kind='bar')
plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.show()
```

A small number of sub-categories contribute significantly to total sales, making them key drivers of business performance.

10. Multiple Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.histplot(df['Sales'], ax=axes[0,0])
axes[0,0].set_title('Sales Distribution')

sns.boxplot(y=df['Profit'], ax=axes[0,1])
axes[0,1].set_title('Profit Distribution')

sns.scatterplot(x='Sales', y='Profit', data=df, ax=axes[1,0])
axes[1,0].set_title('Sales vs Profit')

sns.countplot(x='Category', data=df, ax=axes[1,1])
axes[1,1].set_title('Orders by Category')

plt.tight_layout()
plt.show()
```

The dashboard provides a holistic view of sales, profit, customer behavior, and operational performance, helping identify key trends and business opportunities.
